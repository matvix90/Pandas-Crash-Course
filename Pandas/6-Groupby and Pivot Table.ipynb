```python
import pandas as pd
```

```python
def sp500():
    url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
    html = pd.read_html(url, header = 0)
    data = html[0]
    return data
df = sp500()
df
```

```python
df.sort_values(by = 'GICS Sector')
```

```python
df.sort_values(by = ['GICS Sector', 'GICS Sub-Industry'], ascending = False)
```

```python
df.groupby( by = 'GICS Sector').first()
```

```python
df.groupby( by = 'GICS Sector').sum()['CIK'].sort_values(ascending=False)
```

```python
pd.pivot_table(df, values='CIK', index=['GICS Sector', 'Symbol'])
```
